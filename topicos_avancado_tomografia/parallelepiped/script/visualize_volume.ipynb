```python
import k3d
import numpy as np
from utils.plots import plot_projection_mosaic
```

# Visualizar volume reconstruído

## Carregar dados

```python
data = np.load('../output/volume.npz')
volume = data['volume']
# volume = volume.transpose(0, 1, 2)  # Reorient the volume if necessary

print(f"Data keys: {data.files}")
print(f'Volume shape: {volume.shape}')
print(f'Volume dtype: {volume.dtype}')
print(f'Volume min value: {volume.min()}')
print(f'Volume max value: {volume.max()}')
```

## Visualizar fatias do volume reconstruído

```python
plot_projection_mosaic(volume, num_projections=10, fontsize=14, projection_titles='Fatia')
```

```python
volume_small = volume[::4, ::4, ::4]
vol_norm = (volume_small - volume_small.min()) / (volume_small.max() - volume_small.min())

plot = k3d.plot()
vol_obj = k3d.volume(vol_norm.astype(np.float32),
                     color_map=k3d.colormaps.matplotlib_color_maps.gray,
                     alpha_coef=30)
plot += vol_obj
plot.display()
```
