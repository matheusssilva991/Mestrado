```python
import numpy as np
import matplotlib.pyplot as plt

from utils import (
    min_tnorma,
    prod_tnorma,
    max_tconorma,
    lukasiewicz_tconorma,
    sugeno_negation,
    zadeh_negation,
    young_membership,
    adult_membership,
)
```

# Tarefa Funções de pertinência Idade

```python
age = np.linspace(0, 75, 1000)
mi_young = [young_membership(a) for a in age]
mi_adult = [adult_membership(a) for a in age]

plt.figure(figsize=(10, 6))
plt.plot(age, mi_young, label="Young", color="blue")
plt.plot(age, mi_adult, label="Adult", color="orange")
plt.title("Membership Functions for Age")
plt.xlabel("Age")
plt.ylabel("Membership Degree")
plt.legend()
plt.grid()
plt.show()
```

## 1. T-norma

### Min

```python
min_tornorma_values = [min_tnorma(y, a) for y, a in zip(mi_young, mi_adult)]

figura = plt.figure(figsize=(10, 6))
plt.plot(age, min_tornorma_values, label="Min T-norm", color="green")
plt.title("Min T-norm of Young and Adult Memberships")
plt.xlabel("Age")
plt.ylabel("Membership Degree")
plt.legend()
plt.grid()
plt.show()
```

### Prod

```python
prod_tnorma_values = [prod_tnorma(y, a) for y, a in zip(mi_young, mi_adult)]

figura = plt.figure(figsize=(10, 6))
plt.plot(age, prod_tnorma_values, label="Product T-norm", color="red")
plt.title("Product T-norm of Young and Adult Memberships")
plt.xlabel("Age")
plt.ylabel("Membership Degree")
plt.legend()
plt.grid()
plt.show()
```

## 2. T-conorma

### Max

```python
max_tconorma_values = [max_tconorma(y, a) for y, a in zip(mi_young, mi_adult)]

figura = plt.figure(figsize=(10, 6))
plt.plot(age, max_tconorma_values, label="Max T-conorm", color="purple")
plt.title("Max T-conorm of Young and Adult Memberships")
plt.xlabel("Age")
plt.ylabel("Membership Degree")
plt.legend()
plt.grid()
plt.show()
```

### Lukasiewicz

```python
lukasiewicz_tconorma_values = [lukasiewicz_tconorma(y, a) for y, a in zip(mi_young, mi_adult)]

figura = plt.figure(figsize=(10, 6))
plt.plot(age, lukasiewicz_tconorma_values, label="Lukasiewicz T-conorm", color="brown")
plt.title("Lukasiewicz T-conorm of Young and Adult Memberships")
plt.xlabel("Age")
plt.ylabel("Membership Degree")
plt.legend()
plt.grid()
plt.show()
```

## 3. Negação

### Sugeno

```python
sugeno_negation_values = [sugeno_negation(y) for y in mi_young]
figura = plt.figure(figsize=(10, 6))
plt.plot(age, sugeno_negation_values, label="Sugeno Negation of Young", color="cyan")
plt.title("Sugeno Negation of Young Membership")
plt.xlabel("Age")
plt.ylabel("Membership Degree")
plt.legend()
plt.grid()
plt.show()
```
